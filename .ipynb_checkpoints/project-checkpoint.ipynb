```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("exoplanet_data.csv", comment="#")

# classifies a planet based on temperature
def classify_temperature(temp):
    if pd.isna(temp):
        return "Unknown"
    elif temp < 200:
        return "Too Cold"
    elif temp > 320:
        return "Too Hot"
    else:
        return "Just Right"

# finds potentially habitable planets based on radius + equilibrium temperature
def find_habitable(radius_min, radius_max, temp_min, temp_max):
    habitable = []

    for _, row in data.iterrows():
        radius = row["pl_rade"]
        temp = row["pl_eqt"]

        if pd.notna(radius) and pd.notna(temp):
            if radius_min <= radius <= radius_max and temp_min <= temp <= temp_max:
                habitable.append({
                    "Name": row["pl_name"],
                    "Radius (Earth radii)": radius,
                    "Temperature (K)": temp,
                    "Category": classify_temperature(temp)
                })
    return habitable

# prints results nicely
def print_planets(planets):
    if not planets:
        print("\nNo planets found with those criteria.\n")
        return

    print(f"\nFound {len(planets)} potentially habitable planets:\n")
    for p in planets:
        print(f"{p['Name']} — {p['Radius (Earth radii)']} Earth radii, "
              f"{p['Temperature (K)']} K ({p['Category']})")


# plot 1 (habitable planets)
def plot_planets(planets):
    if not planets:
        print("No data to plot.")
        return

    df = pd.DataFrame(planets)

    plt.figure(figsize=(8, 6))
    plt.scatter(df["Temperature (K)"], df["Radius (Earth radii)"], 
                color="green", label="Habitable Candidates")

    plt.title("Potentially Habitable Exoplanets")
    plt.xlabel("Temperature (K)")
    plt.ylabel("Radius (Earth Radii)")
    plt.grid(True)
    plt.legend()
    plt.show()

# plot 2 (histogram of all exoplanet temperatures)
def plot_temperature_histogram():
    temps = data["pl_eqt"].dropna()

    plt.figure(figsize=(8, 6))
    plt.hist(temps, bins=40)
    plt.title("Distribution of All Exoplanet Temperatures")
    plt.xlabel("Temperature (K)")
    plt.ylabel("Count")
    plt.grid(True)
    plt.show()

# plot 3 (radius vs temperature for all planets)
def plot_all_radius_vs_temp():
    df = data[["pl_rade", "pl_eqt"]].dropna()

    plt.figure(figsize=(8, 6))
    plt.scatter(df["pl_eqt"], df["pl_rade"], alpha=0.4)
    plt.title("All Exoplanets: Radius vs. Temperature")
    plt.xlabel("Temperature (K)")
    plt.ylabel("Radius (Earth radii)")
    plt.grid(True)
    plt.show()

# main interactive loop
while True:
    print("\n=== Exoplanet Habitability Finder ===")
    print("Enter custom limits or press Enter to use default values.\n")

    try:
        radius_min = float(input("Minimum radius (default 0.5): ") or 0.5)
        radius_max = float(input("Maximum radius (default 2.0): ") or 2.0)
        temp_min = float(input("Minimum temperature (default 200): ") or 200)
        temp_max = float(input("Maximum temperature (default 320): ") or 320)
    except ValueError:
        print("\nPlease enter numbers only.\n")
        continue

    planets = find_habitable(radius_min, radius_max, temp_min, temp_max)
    print_planets(planets)

# shows the main plot
    if planets:
        view_plot = input("\nShow plot of these planets? (y/n): ").strip().lower()
        if view_plot == "y":
            plot_planets(planets)

# supporting plots
    print("\n--- Additional Plots ---")
    print("1. Histogram of all exoplanet temperatures")
    print("2. Radius vs temperature for all exoplanets")
    print("3. Skip")

    choice = input("Choose a plot (1/2/3): ").strip()

    if choice == "1":
        plot_temperature_histogram()
    elif choice == "2":
        plot_all_radius_vs_temp()

    again = input("\nSearch again? (y/n): ").strip().lower()
    if again != "y":
        print("\nGoodbye!")
        break
```
