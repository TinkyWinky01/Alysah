```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../../Alysah/homework8/exoplanet data.csv", comment="#")
print("Shape:", df.shape)
print("Rows:", len(df))
print("Columns:", len(df.columns))
print("Column titles:", df.columns.tolist())
df_small = df[["pl_name", "hostname", "discoverymethod", "st_spectype", "pl_eqt", "pl_bmassj"]]
df_small = df_small.dropna()
print("Shape:", df_small.shape)
print("Rows:", len(df_small))
print("Columns:", len(df_small.columns))
print(df_small.head())
print(df_small.tail())
print("Column titles:", df_small.columns.tolist())
```

```python
import numpy as np
import seaborn as sns
print(df_small["discoverymethod"].value_counts())
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_small, x="pl_bmassj", y="pl_eqt", hue="discoverymethod")
plt.title("Planet Mass vs Equilibrium Temperature")
plt.xlabel("Planet Mass (Jupiter Masses)")
plt.ylabel("Equilibrium Temperature (K)")
plt.legend()
plt.show()
plt.figure(figsize=(8,6))
sns.scatterplot(x=np.log(df_small["pl_bmassj"]), y=np.log(df_small["pl_eqt"]), hue=df_small["discoverymethod"])
plt.title("Planet Mass vs Equilibrium Temperature (Log Scale)")
plt.xlabel("Log Planet Mass")
plt.ylabel("Log Equilibrium Temperature")
plt.legend()
plt.show()
coldest = df_small[df_small["pl_bmassj"] > 5].sort_values("pl_eqt").iloc[0]
print(coldest)
plt.figure(figsize=(8,6))
sns.scatterplot(x=np.log(df_small["pl_bmassj"]), y=np.log(df_small["pl_eqt"]), hue=df_small["discoverymethod"])
plt.scatter(np.log(coldest["pl_bmassj"]), np.log(coldest["pl_eqt"]), color="red", marker="*", s=300, label=coldest["pl_name"])
plt.title("Planet Mass vs Equilibrium Temperature (Log Scale)")
plt.xlabel("Log Planet Mass")
plt.ylabel("Log Equilibrium Temperature")
plt.legend()
plt.show()
```
