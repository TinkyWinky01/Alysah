```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

df = pd.read_csv("C:/Users/ander/python_decal_fa25/course_assignments/homework9/noisy_curvefit_data.csv")

x = df["x"].values
y_noise_1 = df["y_noise_1"].values

def model(x, A, k):
    return A * np.exp(-k * x)

p0 = [1, 1]
popt, pcov = curve_fit(model, x, y_noise_1, p0=p0)

A_fit, k_fit = popt

print(f"Fit: y = {A_fit:.2f} * e^(-{k_fit:.2f} * x)")

residuals = y_noise_1 - model(x, A_fit, k_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_reduced = chi2 / dof
print(f"Reduced chi-squared: {chi2_reduced:.2f}")

plt.figure(figsize=(8,6))
plt.scatter(x, y_noise_1, color="gray", label="Noisy data (level 1)")
plt.plot(x, model(x, A_fit, k_fit), color="red", label="Fitted model")
plt.plot(x, model(x, 5, 0.8), color="black", linestyle="--", label="True model")

plt.title(f"Noise Level 1 Fit (_reduced = {chi2_reduced:.2f})")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
def fit_model(x, y):
    p0 = [1, 1]
    popt, pcov = curve_fit(model, x, y, p0=p0)
    A_fit, k_fit = popt
    residuals = y - model(x, A_fit, k_fit)
    chi2 = np.sum(residuals**2)
    dof = len(x) - len(popt)
    chi2_reduced = chi2 / dof
    return A_fit, k_fit, chi2_reduced
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
noise_cols = ["y_noise_2", "y_noise_3", "y_noise_4", "y_noise_5"]

for ax, col in zip(axes.flatten(), noise_cols):
    y = df[col].values
    A_fit, k_fit, chi2_reduced = fit_model(x, y)
    
    ax.scatter(x, y, color="gray", label="Noisy data")
    
    ax.plot(x, model(x, A_fit, k_fit), color="red", label="Fitted model")
    
    ax.plot(x, model(x, 5, 0.8), color="black", linestyle="--", label="True model")
    
    ax.set_title(f"{col}: _reduced = {chi2_reduced:.2f}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()

plt.tight_layout()
plt.show()
#What happens to the red and black dotted lines as you increase the noise level? Why?: The red fitted line drifts farther away from the black dotted line because the noise makes the trend harder to see.
#What happens to reduced chi-squared as noise increases?: It gets bigger. More noise means the data points stray further from the fitted curve, so the fit looks worse and x^2(reduced) goes up.

results = []
for i, col in enumerate(["y_noise_1", "y_noise_2", "y_noise_3", "y_noise_4", "y_noise_5"], start=1):
    y = df[col].values
    A_fit, k_fit, chi2_reduced = fit_model(x, y)
    results.append([i, A_fit, k_fit, chi2_reduced])

results_df = pd.DataFrame(results, columns=["Noise level", "A_fit", "k_fit", "Reduced chi-squared"])
print(results_df)

df.to_csv("C:/Users/ander/python_decal_fa25/course_assignments/homework9/noisy_parameters.csv", index=False)
```
