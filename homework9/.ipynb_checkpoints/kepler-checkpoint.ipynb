```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/ander/python_decal_fa25/course_assignments/homework9/solar_system.csv")

print("Shape:", df.shape)
print("Rows:", len(df))
print("Columns:", len(df.columns))

print("\nFirst five rows:")
print(df.head())

print("\nLast five rows:")
print(df.tail())

print("\nColumn names:")
print(df.columns.tolist())

print("\nRow names (index):")
print(df.index.tolist())
```

```python
from scipy.optimize import curve_fit

df_T = df.set_index("Attribute").T
print(df_T.head())

orbital_period = df_T["Orbital Period (days)"].astype(float)
perihelion     = df_T["Perihelion (10^6 km)"].astype(float)
aphelion       = df_T["Aphelion (10^6 km)"].astype(float)
print("Orbital Periods (days):", orbital_period.values)
print("Perihelion (10^6 km):", perihelion.values)
print("Aphelion (10^6 km):", aphelion.values)

semi_major_axis = (perihelion + aphelion) / 2
df_T["Semi-major axis (10^6 km)"] = semi_major_axis
print("Semi-major axis (10^6 km):", semi_major_axis.values)


#Perhelion: The closest point in a planet's orbit to the Sun.
#Aphelion: The farthest point in a planet's orbit from the Sun.
#Semi-major axis: Half the diameter of an elliptical orbit.

plt.figure(figsize=(8,6))
plt.scatter(df_T["Semi-major axis (10^6 km)"], df_T["Orbital Period (days)"], 
            color="blue", label="Planets")
plt.xlabel("Semi-major axis (10^6 km)")
plt.ylabel("Orbital Period (days)")
plt.title("Kepler's Third Law: Semi-major axis vs Orbital Period")
plt.legend()
plt.show()


def keplers_third_law(a, m):
    return m * a ** 1.5

T_years = df_T["Orbital Period (days)"].astype(float) / 365.25
a_AU = df_T["Semi-major axis (10^6 km)"].astype(float) / 149.6

from scipy.optimize import curve_fit

popt, pcov = curve_fit(keplers_third_law, a_AU, T_years, p0=[1])
m_fit = popt[0]

print(f"Fit: T = {m_fit:.2f} * a^(3/2)")

residuals = T_years - keplers_third_law(a_AU, m_fit)
chi2 = np.sum(residuals**2)
dof = len(a_AU) - len(popt)
chi2_reduced = chi2 / dof
print(f"Reduced chi-squared: {chi2_reduced:.2f}")

plt.figure(figsize=(8,6))
plt.scatter(a_AU, T_years, color="blue", label="Planets")
plt.plot(a_AU, keplers_third_law(a_AU, m_fit), color="red", label="Fitted model")
plt.plot(a_AU, keplers_third_law(a_AU, 1), color="black", linestyle="--", label="True law (m=1)")

plt.xlabel("Semi-major axis (AU)")
plt.ylabel("Orbital Period (years)")
plt.title(f"Kepler's Third Law Fit (_reduced = {chi2_reduced:.2f})")
plt.legend()
plt.show()

df_T.to_csv("C:/Users/ander/python_decal_fa25/course_assignments/homework9/kepler.csv", index=True)
```
