```python
#REMINDER FOR MYSELF; TRY TO SEPARATE EVERYTHING WITH NOTES CAUSE YOU GOT CONFUSED LAST TIME
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np

path = "C:/Users/ander/python_decal_fa25/course_assignments/homework10"
print(os.listdir(path))

filenames = [
    "M13_0001.fits",
    "M13_0002.fits",
    "M13_0003.fits",
    "M13_0004.fits",
    "M13_0005.fits"
]

# load fits data
data_list = []
headers = []

for fname in filenames:
    with fits.open(os.path.join(path, fname)) as hdul:
        headers.append(hdul[0].header)
        data_list.append(hdul[0].data)

print("Loaded {len(data_list)} FITS files")

# individual scaling
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for i, ax in enumerate(axes):
    ax.imshow(data_list[i], origin="lower", cmap="gray")
    ax.set_title(f"M13 {i+1:04d}")
    ax.axis("off")
plt.tight_layout()
plt.show()

# consistent scaling
vmin = min([np.min(d) for d in data_list])
vmax = max([np.max(d) for d in data_list])

fig, axes = plt.subplots(1, 5, figsize=(20, 4), constrained_layout=True)
for i, ax in enumerate(axes):
    im = ax.imshow(data_list[i], origin="lower", cmap="gray", vmin=vmin, vmax=vmax)
    ax.set_title(f"M13 {i+1:04d}")
    ax.axis("off")

fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.6, label="Pixel intensity")
plt.show()

# average and median
stacked_avg = np.mean(data_list, axis=0)
stacked_med = np.median(data_list, axis=0)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

im1 = axes[0].imshow(stacked_avg, origin="lower", cmap="gray", vmin=vmin, vmax=vmax)
axes[0].set_title("Average Stacked Image")

im2 = axes[1].imshow(stacked_med, origin="lower", cmap="gray", vmin=vmin, vmax=vmax)
axes[1].set_title("Median Stacked Image")

fig.colorbar(im1, ax=axes.ravel().tolist(), shrink=0.6, label="Pixel intensity")
plt.tight_layout()
plt.show()

# plot of histograms
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i, ax in enumerate(axes):
    data = data_list[i]
    ax.hist(data.ravel(), bins=50, color="steelblue", alpha=0.7)
    ax.set_title(f"M13 {i+1:04d}")
    ax.set_xlabel("Pixel Intensity")
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()

#(a) The x-axis values are the pixel intensity values. The y-axis values are the frequency.
#(b) bins=50 divides the range of pixel intensities into 50 equal bins.
#(c) Most data centered for pixel intensity is located near low intensities.
#(d) The data isn't that spread out, as it appears close together.
#(e) The median is close to the background level, while the mean is higher due to bright stars.

combined_image = np.median(data_list, axis=0)
median_val = np.median(combined_image)
print("Median pixel intensity of combined image: {median_val:.2f}")
plt.figure(figsize=(8,6))
plt.hist(combined_image.ravel(), bins=50, color="steelblue", alpha=0.7)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.title("Histogram of combined image")
plt.show()
hist_vals, bin_edges = np.histogram(combined_image.ravel(), bins=50)
vmin = np.percentile(combined_image, 5) # The image was blank so I adjusted the vmin and vmax
vmax = vmin + 100

print("Using vmin={vmin:.2f}, vmax={vmax:.2f}")
plt.figure(figsize=(6,6))
plt.imshow(combined_image, origin="lower", cmap="gray", vmin=vmin, vmax=vmax)
plt.colorbar(label="Pixel Intensity")
plt.title("Median combined image of M13")
plt.show()

fits.writeto("combined.fits", combined_image, overwrite=True)
print("combined image saved as combined.fits")
```
