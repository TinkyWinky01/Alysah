```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
print("Data shape:", data.shape)
magnitude = data[:, 0]     # Brightness
temperature = data[:, 1]   # Temperature in Kelvin
plt.figure(figsize=(10, 6))
plt.scatter(temperature, magnitude, c=temperature, cmap='plasma', s=10)

plt.gca().invert_xaxis()  # Hotter stars on the left
plt.gca().invert_yaxis()  # Brighter stars at the top

plt.xlabel("Temperature (K)", fontsize=14)
plt.ylabel("Magnitude", fontsize=14)
plt.title("Hertzsprung-Russell Diagram", fontsize=16)
plt.colorbar(label="Temperature (K)")
plt.grid(True)
plt.show()
```

```python
import numpy as np

data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
print("Data shape:", data.shape)
magnitude = data[:, 0]
temperature = data[:, 1]
print("Minimum magnitude:", np.min(magnitude))
print("Maximum magnitude:", np.max(magnitude))
print("Minimum temperature:", np.min(temperature))
print("Maximum temperature:", np.max(temperature))
brightest_index = np.argmin(magnitude)
dimmest_index = np.argmax(magnitude)

print("Brightest star  Magnitude:", magnitude[brightest_index], "Temperature:", temperature[brightest_index])
print("Dimmest star  Magnitude:", magnitude[dimmest_index], "Temperature:", temperature[dimmest_index])
```

```python
import numpy as np
import matplotlib.pyplot as plt
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
plt.figure(figsize=(8, 6))
plt.scatter(log_temperature, log_magnitude, label="Stars", alpha=0.6)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
brightest_idx = np.argmin(magnitude)
dimmest_idx = np.argmax(magnitude)
hottest_idx = np.argmax(temperature)
coldest_idx = np.argmin(temperature)

plt.scatter(log_temperature[brightest_idx], log_magnitude[brightest_idx],
            color='black', marker='*', s=300, label='Brightest Star')

plt.scatter(log_temperature[dimmest_idx], log_magnitude[dimmest_idx],
            color='purple', marker='*', s=300, label='Dimmest Star')

plt.scatter(log_temperature[hottest_idx], log_magnitude[hottest_idx],
            color='green', marker='*', s=300, label='Hottest Star')

plt.scatter(log_temperature[coldest_idx], log_magnitude[coldest_idx],
            color='orange', marker='*', s=300, label='Coldest Star')
plt.legend()
plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")

plt.grid(True)
plt.show()
```
